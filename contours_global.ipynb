```python
import cv2
import matplotlib.pyplot as plt
import os #pour avoir accès aux photos dans le dossier
```

```python
source_folder = "images"
destination_folder = "contours"
numéro_page = 0

for filename in os.listdir(source_folder): #pour chaue page dans le dossier image
    source_path = os.path.join(source_folder, filename)
    dest_path = os.path.join(destination_folder, filename)

    img = cv2.imread("images/" + filename) #on va lire toutes les images
    page = img.copy()
    #on applique le code pour 1 image, à tout le dossier
    gray = cv2.cvtColor(page, cv2.COLOR_BGR2GRAY)
    flou = cv2.blur(gray,(10, 10), 3)
    thresh = cv2.adaptiveThreshold(flou, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY_INV, 11, 2)

    kerneldil = cv2.getStructuringElement(cv2.MORPH_RECT, (7,6)) 
    kernelero = cv2.getStructuringElement(cv2.MORPH_RECT, (1,1)) 

    morph = cv2.erode(thresh, kernelero, iterations = 1 )
    morph2 = cv2.dilate(morph, kerneldil, iterations=1)

    contours, hierarchy = cv2.findContours(morph2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w > 50 and h > 50:
            cv2.rectangle(page, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    cv2.imwrite(dest_path, page)
    print(f"[OK] {filename} enregistrée.")
```
