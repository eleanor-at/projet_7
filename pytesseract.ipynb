```python
import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
import Levenshtein
```

```python
page21 = cv2.imread('page21.jpg')
page127 = cv2.imread("page127.jpg")
page589 = cv2.imread("page589.jpg")
page568 = cv2.imread("image-0568.png")
page70 = cv2.imread("image-0070.png")
```

```python
page595 = cv2.imread("image-0595.png")
page = page21.copy()


gray = cv2.cvtColor(page, cv2.COLOR_BGR2GRAY)
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

texte = pytesseract.image_to_string(gray, lang='lat') 

print(texte)
```
